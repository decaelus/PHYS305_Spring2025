### 1.

We can relate the path length before reflection $s_1$ and the path length after reflection $s_2$ to the total path length:
$$
s = s_1 + s_2 = \frac{x}{\sin \theta_1} + \frac{x_0 - x}{\sin \theta_2}.
$$

Chromey suggests setting $\frac{ds}{dx} = 0$ to find an extremum in the length. This constraint involves Fermat's principle since the index of refraction along both lengths is the same, and therefore the speed of light is the same. If the indices of refraction were different (i.e., as in problem 2), then the relationship between $\theta_1$ and $\theta_2$ would explicitly involve the indices.

$$
\frac{ds}{dx} = \frac{1}{\sin \theta_1} - \frac{1}{\sin \theta_2} = 0 \Rightarrow \boxed{ \sin \theta_1 = \sin \theta_2 }.
$$

### 2.

For a fixed vertical distance $y$, the path length through the first medium $s_1$ is
$$
s_1 = \sqrt{x^2 + y^2}.
$$
Similarly for the length through the second medium $s_2$:
$$
s_2 = \sqrt{\left( X_0 - x \right)^2 + y_2^2}.
$$

The velocity through the first medium is $v_1 = c/n_1$ and through the second $v_2 = c/n_2$. Therefore the total time $t$ to traverse both media is
$$
t = \frac{s_1}{v_1} + \frac{s_2}{v_2} = \frac{n_1 \sqrt{x^2 + y^2}}{c} + \frac{n_2 \sqrt{\left( X_0 - x \right)^2 + y_2^2}}{c}.
$$

As suggested by the problem, let's find the relationship that satisfies $\frac{dt}{dx} = 0$:
$$
0 = \frac{n_1 x}{c\sqrt{x^2 + y^2}} - \frac{n_2 \left( X_0 - x \right)}{c\sqrt{\left( X_0 - x \right)^2 + y_2^2}}.
$$

We can multiply both sides by $c$ and re-arrange:
$$
\frac{n_1 x}{\sqrt{x^2 + y^2}} = \frac{n_2 \left( X_0 - x \right)}{\sqrt{\left( X_0 - x \right)^2 + y_2^2}}.
$$

And we can see that $\sin \theta_1 = \frac{x}{\sqrt{x^2 + y^2}}$, while $\sin \theta_2 = \frac{\left( X_0 - x \right)}{c\sqrt{\left( X_0 - x \right)^2 + y_2^2}}$, giving
$$
\boxed{ n_1 \sin \theta_1 = n_2 \sin \theta_2 }.
$$

### 7.

We can start by calculating the index of refraction at each wavelength $\lambda$.

At $\lambda = 350\,{\rm nm}$, $n = 1.52 + \left( 0.00436 \,{\rm \mu m}^2 \right) \left(0.35\,{\rm \mu m}\right)^{-2} = 1.56$.

At $\lambda = 600\,{\rm nm}$, $n = 1.52 + \left( 0.00436 \,{\rm \mu m}^2 \right) \left(0.6\,{\rm \mu m}\right)^{-2} = 1.53$.

Next, we can use Equation 5.26 (p. 13) to calculate the angle of minimum deflection $\alpha$ for each wavelength.

At minimum deflection, $\alpha = \theta_0$, and
$$
\alpha = \theta_0 = 2\sin^{-1} \left( n \sin \left( A/2 \right) \right) - A,
$$

For $\lambda = 350\,{\rm nm}$,
$$
\theta_0 = 2 \sin^{-1} \left( 1.56 \sin \left( 15^\circ \right) \right) - 30^\circ = 17.6^\circ,
$$

and for $\lambda = 600\,{\rm nm}$,
$$
\theta_0 = 2 \sin^{-1} \left( 1.53 \sin \left( 15^\circ \right) \right) - 30^\circ = 16.7^\circ.
$$

We can roughly estimate $d\theta/d\lambda$ as 
$$
\frac{d\theta}{d\lambda} \approx \frac{\Delta \theta_0}{\Delta \lambda} = \frac{16.7^\circ - 17.6^\circ}{600\,{\rm nm} - 350\,{\rm nm}} = -0.0628{\rm \mu m}^{-1}
$$

Next, let's use Eqn 5.29 to estimate deflections at each wavelength. We'll need first to evaluate $g$ at $\alpha$. Strictly, we'd need to evaluate it at each value of $\alpha$, but they are both so close, let's just take one value:
$$
g = -4 \left( 0.00436\,{\rm \mu m}^2 \right) \frac{\sin \left( 30^\circ/2 \right)}{\cos \left( 17^\circ \right)} = 0.00459\,{\rm \mu m}^2\\
$$

For $\lambda = 600\,{\rm nm}$,
$$
\frac{\partial \theta}{\partial \lambda} = \frac{0.00459\,{\rm \mu m}^2}{\left( 0.6\,{\rm \mu m}\right)^3} = 0.02.
$$

For $\lambda = 350\,{\rm nm}$,
$$
\frac{\partial \theta}{\partial \lambda} = \frac{0.00459\,{\rm \mu m}^2}{\left( 0.35\,{\rm \mu m}\right)^3} = 0.11.
$$

Taking the average: $0.5\left( 0.02 + 0.11 \right) = 0.064\,{\rm \mu m}^{1}$, which is just about $0.0628\,{\rm \mu m}^{1}$.

### 8.

With a focal ratio $\mathscr{R} = f/D = 6.0$ and an aperture $D = 1.0\,{\rm m}$, we can calculate the focal length $f = D \times \mathscr{R} = 6.0\,{\rm m}$.

That focal length translates into an image scale $s = 1/f = 1\,{\rm radians}/6.0\,{\rm m} = 0.17\,{\rm radians\ m^{-1}}$.

Each pixel is $18\,{\rm \mu m}$ across, so the pixel scale is 
$$
s_{\rm p} = s d = \left( 0.17\,{\rm radians\ m^{-1}} \right) \times \left( 18\times10^{-6}\,{\rm m}\right) \times \left( \frac{180^\circ}{\pi\,{\rm radians}} \right) \left( \frac{60'}{1^\circ} \right) = \boxed{ 0.01' }.
$$

With a CCD 1024 pixels across, the camera frame is $F = \left( 1024 \right)\times \left( 18\,{\rm \mu m} \right) = 18\,{\rm mm}$ across. Using the image scale, we get a frame width $\Delta = F \times s = \left( 0.018\,{\rm m} \right) \times \left( 0.17\,{\rm radians\ m^{-1}} \right) = 0.003\,{\rm radians} \times \left( \frac{180^\circ}{\pi\,{\rm radians}} \right) \left( \frac{60'}{1^\circ} \right) = \boxed{ 11' }$, about a third the angular width of the Moon.

### 9.

The diffraction limit $\alpha$ is given by:
$$
\alpha = \frac{1.22\ \lambda}{D},
$$
where $\lambda$ is the wavelength and $D$ is the telescope diameter.

With $\lambda = 2\,{\rm \mu m}$ and $D = 24\,{\rm m}$,
$$
\alpha = \frac{1.22\left( 2\times10^{-6}\,{\rm m} \right)}{24\,{\rm m}} = 10^{-7}\, {\rm radians}.
$$

And according to the problem, we expect the telescope to have a resolution twice that, so $2\times10^{-7}\, {\rm radians}$.

We are also told that the system is an $f/3$ system, meaning $f/D = 3 \Rightarrow f = 3\times \left( 24\,{\rm m}\right) = 72\,{\rm m}$.

So we can use that focal length to calculate the image scale:
$$
s = \frac{1}{f} = 0.01\,{\rm radians\ m^{-1}}.
$$

So each meter of detector can resolve 0.01 radians. We can combine the resolution and image scale:
$$
\frac{2\times10^{-7}\,{\rm radians}}{0.01\,{\rm radians\ m^{-1}}} = 2\times10^{-5}\,{\rm m} = 20\,{\rm \mu m}.
$$

That number represents the required spacing of pixels to resolve each point source with a single pixel, but we're told we want 2 pixels across, so we want pixels half as big, i.e., $\boxed{ 10\,{\rm \mu m}}$.

### 11.

We're told Mars emits a flux $F = 1.0\times10^{-7}\,{\rm W\ m^{-2}}$. We can calculate the power $E$ captured by a telescope with diameter $D$ with an exposure time $t$ using:
$$
E = \pi \left( D/2 \right)^2 t F.
$$

The total energy collected by the telescopes is then
$$
E_{\rm Albert} = \pi \left( 0.3\,{\rm m} / 2 \right)^2 \left( 100\,{\rm seconds} \right) \left( 1.0\times10^{-7}\,{\rm W\ m^{-2}} \right) = 7.1\times10^{-7}\, {\rm J},\ \\
E_{\rm Berta} = \pi \left( 30\,{\rm m} / 2 \right)^2 \left( 100\,{\rm seconds} \right) \left( 1.0\times10^{-7}\,{\rm W\ m^{-2}} \right) = 7.1\times10^{-3}\, {\rm J}
$$

And so now, for this problem, we need to figure out how many pixels that energy is spread over within each telescope. 

We need to calculate the image scale $s$ for each telescope. To do that, we'll first need to calculate the focal lengths $f$ from the f-numbers:
$$
f_{\rm Albert} = 8\times \left( 0.3\,{\rm m} \right) = 2.4\,{\rm m},\ \\
f_{\rm Bertha} = 4\times \left( 30\,{\rm m} \right) = 120\,{\rm m}.
$$

We then calculate the image scales:
$$
s_{\rm Albert} = \frac{1}{2.4\,{\rm m}} = 0.42\,{\rm radians\ m^{-2}},\ \\
s_{\rm Bertha} = \frac{1}{120\,{\rm m}} = 0.0083\,{\rm radians\ m^{-2}}.
$$

Then we'll use the pixel spacing to determine the pixel scale $s_{\rm d}$ for each telescope (and convert that to arcseconds since that's the units used for Mars' angular size):
$$
s_{\rm d, Albert} = \left( 0.42\,{\rm radians\ m^{-2}} \right) \left( 25\,{\rm \mu m} \right) = 10^{-5}\,{\rm radians} \times \left( \frac{180^\circ}{\pi\,{\rm radians}} \right) \times \left( \frac{60\times60\,{\rm arcseconds}}{1^\circ} \right) = 2.1''\\
s_{\rm d, Bertha} = \left( 0.0083\,{\rm radians\ m^{-2}} \right) \left( 25\,{\rm \mu m} \right) = 2.1\times10^{-7}\,{\rm radians}\times \left( \frac{180^\circ}{\pi\,{\rm radians}} \right) \times \left( \frac{60\times60\,{\rm arcseconds}}{1^\circ} \right) = 0.04''
$$

Finally, we need to divide up the image of Mars across all the pixels. We're told the image has an angular diameter of $15''$, which corresponds to an angular area of about $ \left( 15''/2 \right)^2 = 56\,{\rm arcseconds^2}$ (ignoring the fact that image is circular).

Each telescope's pixel have an angular area $A$:
$$
A_{\rm Albert} = \left( 2.1'' \right)^2 = 4.4\,{\rm arcseconds^2},\ \\
A_{\rm Bertha} = \left( 0.04'' \right)^2 = 0.0016\,{\rm arcseconds^2}.
$$

So Albert's telescope will spread Mars over $56/4.4 \approx 13$ pixels, while Bertha's scope will spread it over $56/0.0016 = 35000.$

The total energy collected by each telescope will be spread over all those pixels, so the energy accumulated per pixel for each scope is
$$
\boxed{ E_{\rm Albert}/N_{\rm Albert} = 7.1\times10^{-7}\, {\rm J}/13 = 5.4\times10^{-8}\, {\rm J},\ \\
E_{\rm Bertha}/N_{\rm Bertha} = 7.1\times10^{-3}\, {\rm J}/35000 = 2.0\times10^{-7}\, {\rm J} }
$$

### 12.

The diffraction limit $\alpha$ for a telescope of diameter $D$ observing in a wavelength $\lambda$ is given by Equation 5.37:
$$
\alpha = \frac{1.22\ \lambda}{D}.
$$

Hubble has $D = 2.4 \,{\rm m}$, so at $\lambda = 300\, {\rm nm}$, $\alpha = \frac{1.22 \left( 300\times10^{-9}\, {\rm m}\right)}{2.4 \,{\rm m}} = 0.15\,{\rm \mu radians} = 31\,{\rm mas}$. At $\lambda = 2.0\, {\rm \mu m}$, $\alpha = \frac{1.22 \left( 2.0\times10^{-6}\,{\rm m}\right)}{2.4 \,{\rm m}} = 1.0\mu{\rm radians} = 0.2\,{\rm arcsecs}$.

Your pupil is about $4\, {\rm mm}$ in diameter, so a $\lambda = 0.5\,{\rm \mu m}$, $\alpha = \frac{1.22 \left( 4\times10^{-3}\, {\rm m}\right)}{0.5\times10^{-6} \,{\rm m}} = 150\,{\rm \mu radians} = 30\,{\rm arcsecs}$.

A space telescope with $D = 8\,{\rm m}$ is $8/2.4 = 3.3$ times larger, so at the same wavelength ($2.0\, {\rm \mu m}$), it will have three times better angular resolution than Hubble.

A 30-m telescope is 12.5 times larger than Hubble and so would have 12.5-times the angular resolution. Of course, such a telescope on the ground would have to contend with atmospheric turbulence, which would significantly degrade its effective resolution.
